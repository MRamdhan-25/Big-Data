```python
teks = ['Anto', 'adalah', 'anak','yang', 'suka', 'olahraga', 'dan', 'anak', 'yang', 'suka', 'makan']

#fungsi map untuk memetakan kata-kata dari kalimat dari satu kalimat
map_result = [(kata, 1) for kata in teks]

#fungsi reduce untuk menggabungkan kata yang sama dari satu kalimat
reduce_result = {}
for key, value in map_result:
    reduce_result[key] = reduce_result.get(key, 0) + value

print("Jumlah kata: ", reduce_result)
```

```python
# Map function
def map_function(text):
    for word in text.split():
        yield (word, 1)

# Reduce function
from collections import defaultdict

def reduce_function(pairs):
    result = defaultdict(int)
    for word, count in pairs:
        result[word] += count
    return result

text = "Paijo adalah anak yang suka tidur dan Paijo anak yang suka menggambar"
map = list(map_function(text))
reduce = reduce_function(map)

for kata, jumlah in reduce.items():
    print(f"{kata}: {jumlah}")
```

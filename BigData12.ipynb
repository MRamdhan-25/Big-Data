```python
# Exercies 1: mengidentifikasi dan menangani data yang hilang
import pandas as pd

# Contoh dataset dengan nilai yang hilang 
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', None],
    'Age' : [20, 30, None, 22, 35],
    'Salary' : [48000, None, 57000, None, 60000]
}

df = pd.DataFrame(data)

# Mengisi nilai yang hilang dan menghapus baris
df['Age'].fillna(df['Age'].mean(), inplace = True)
df['Salary'].fillna(df['Salary'].median(), inplace = True)
df.dropna(subset=['Name'], inplace = True)
print("After Clenaning:\n", df)
```

```python
# Exercies 2: Standarisasi Data Kategori
# Contoh dataset dengan nilai kategori yang tidak konsisten
data = {
    'Product': ['Laptop', 'Laptop', 'Desktop', 'Tablet', 'Tablet'],
    'Category': ['Electronics', 'electronics', 'Electronics', 'Gadgets', 'gadgets']
}
df = pd.DataFrame(data)

#Standarisasi nilai kategori
df['Category'] = df['Category'].str.capitalize()
print("Standarized Data:\n", df)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# 1. Identifikasi Missing Values
print("Nilai yang hilang per Kolom:")
print(df.isnull().sum())

# 2. Normalisasi kolom numerik
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

scaler = MinMaxScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

print("\nData setelah normalisasi:")
print(df.head(6))


# 3. Standardisasi kolom kategorikal & hapus duplikasi
categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    df[col] = df[col].astype(str).str.capitalize()

df = df.drop_duplicates()

print("\nData setelah standardisasi dan penghapusan duplikasi:")
print(df.head())
```
